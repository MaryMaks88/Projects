```python
df_enc['sex'].value_counts()
```

```python
df_enc['capital.gain'] = np.log1p(df_enc['capital.gain'])
```

```python
df_enc['capital.loss'] = np.log1p(df_enc['capital.loss'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

df_enc['hours.per.week'] = scaler.fit_transform(df_enc[['hours.per.week']])
df_enc['education.num'] = scaler.fit_transform(df_enc[['education.num']])
df_enc['age'] = scaler.fit_transform(df_enc[['age']])
```

```python
df_enc['native.country'].value_counts()
```

```python
df_enc['native.country'] = df_enc['native.country'].replace('?', 'Other')
```

```python
df_enc['native.country'].value_counts().head()
```

```python
df_enc['income'].value_counts()
```

```python
df_enc['income'] = df_enc['income'].map({'<=50K': 0, '>50K': 1})
```

```python
df_enc['income'].value_counts()
```

```python
import category_encoders as ce
```

```python
cat_columns = ['workclass', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
```

```python
df_enc.head()
```
